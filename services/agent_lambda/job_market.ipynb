```python
import requests

url = "https://api.theirstack.com/v1/jobs/search"
body = """{
  "page": 0,
  "limit": 25,
  "job_country_code_or": [
    "US"
  ],
  "posted_at_max_age_days": 7
}"""
response = requests.request("POST", url, data = body, headers = {
  "Content-Type": "application/json"
})

print(response.text)
```
